Bildkompression ist eine häufige Anwendung von Clustering-Algorithmen, bei der 
ähnliche Farben zu einem begrenzten Satz von Farben zusammengefasst werden, 
um die Dateigröße zu reduzieren. Der K-Means-Algorithmus eignet sich gut für 
diese Aufgabe, da er Cluster von ähnlichen Farben identifizieren kann.

### Schritte zur Bildkompression mit K-Means:
1. **Daten einlesen:** Lese das Bild in Python ein.
   
2. **Pre-Processing:** Konvertiere das Bild in eine geeignete Datenstruktur. Dazu eignet sich ein Array aus RGB-Werten.

3. **K-Means Clustering:** Wähle die Anzahl an Clustern (K) aus und wende den K-Means-Clustering Algorithmus auf die Daten an. Der Algorithmus weist jedem Pixel im Bild das Cluster zu, dessen Schwerpunkt am nächsten zu den Farbwerten des Pixels liegt.

4. **Rekonstruktion:** Ersetze die Farbwerte jedes Pixels im Bild durch die Farbewerte vom Schwerpunkt des zugewiesenen Clusters.

5. **Evaluation:** Zeige das Original und das komprimierte Bild nebeneinander an und vergleiche. Experimentiere mit verschiedenen Werten für K.

6. **Auswertung der Kompression:** Speichere das komprimierte Bild und berechne das Kompressionsverhältnis. Vergleiche dazu die Dateigröße des Orginalbildes mit der des komprimierten Bildes.

Der Ernstfall ist eingetreten! Nemo ist wieder

bei der Suche nach Nemo zu helfen. Lese das Bild ```bruce.jpg```

**Tip:** Verwenden Sie ```plt.imread```, ```plt.imshow``` und ```plt.imsave```
zum lesen, anzeigen und speichern von Bildern.

````python
# Bild einlesen und anzeigen
bild = plt.imread('bruce.jpg')
plt.imshow(bild)
plt.axis('off')  # Achsen ausblenden
plt.show()
````

### Konvertieren der Bildaten 
Konvertiere das Bild zu einem Numpy-Array aus RGB-Werten.

**Tip**: Es kann helfen das Bild von unit8 zu float32 zu konvertieren und zwischen 0-1 zu skalieren.

````python
# Bilddaten normalisieren und umformen
bildArr = np.array(bild, dtype=np.float32) / 255  # Werte zwischen 0 und 1
original_shape = bildArr.shape
bildArr = bildArr.reshape((-1, 3))  # In ein 2D-Array (Pixelanzahl x 3 Farben) umwandeln
````

### K-Means-Clustering
Implementiere den K-Means-Clustering-Algorithmus in der gegebenen Klasse.

**Methoden:**
* ```__init__()```: Kontruktor.

* ```score()```: Kostenfunktion.

* ```fit()```: Berechnet die Clusterschwerpunkte (centroids).

* ```predict()```: Weist jedem Pixel die Farbe des jeweiligen Schwerpunkts zu.

**Parameter:**
* ```n_clusters```: Anzahl der Cluster.

* ```max_iter```: Maximale Anzahl an Iterationen.

* ```tol```:  Konvergenzkriterium. Die relative Toleranz der Differenz der Clusterschwerpunkte zwischen zwei aufeinanderfolgenden Iterationen.


````python
class KMeans:
    def __init__(self, n_clusters=16, max_iter=100, tol=1e-3, random_seed=42):
        np.random.seed(random_seed)
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.tol = tol
        self.centroids = None

    def fit(self, X):
        h, *dims = X.shape
        self.centroids = X[np.random.choice(h, self.n_clusters, replace=False)]

        for iteration in range(self.max_iter):
            distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)
            cluster_labels = np.argmin(distances, axis=1)

            new_centroids = np.array([
                X[cluster_labels == k].mean(axis=0) if np.any(cluster_labels == k) else self.centroids[k]
                for k in range(self.n_clusters)
            ])

            if np.linalg.norm(new_centroids - self.centroids) < self.tol:
                break

            self.centroids = new_centroids

    def predict(self, X):
        distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)
        return np.argmin(distances, axis=1)

    def score(self, X):
        distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)
        closest_distances = np.min(distances, axis=1)
        return np.sum(closest_distances ** 2)
````

### Rekonstruktion
Ersetze die Farbwerte jedes Pixels im Bild durch die 
Farbewerte vom Schwerpunkt des zugewiesenen Clusters. Nutze dazu die in implementierte
KMeans-Klasse. Die Methode ```fit()``` berechnet die Clusterschwerpunkte, ```predict()```
ersetzt die Farbwerte von jedem Pixel durch die Farbewerte vom nächstgelegenen Clusterschwerpunkt.

**Tip**: Verwende eine geringe Anzahl an Iterationen (z. B. 30) aufgrund der hohen Laufzeit.

````python
# K-Means-Clustering mit K=16
kmeans = KMeans(n_clusters=16, max_iter=30)
kmeans.fit(bildArr)
labels = kmeans.predict(bildArr)

# Rekonstruktion des komprimierten Bildes
compressed_img = kmeans.centroids[labels].reshape(original_shape)

# Anzeige des Original- und komprimierten Bildes
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(bild)
plt.title("Originalbild")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(compressed_img)
plt.title("Komprimiertes Bild (K=16)")
plt.axis('off')

plt.show()
````

### Evaluation

Probiere verschiedene Werte für ```K``` aus. Zeige drei Bilder nebeneinander an:

````python
# Evaluation: Speichere das komprimierte Bild
plt.imsave('compressed_bruce.jpg', compressed_img)

# Lade die Dateigrößen für Original und Komprimierung
import os
original_size = os.path.getsize('bruce.jpg')  # Größe des Originalbilds in Bytes
compressed_size = os.path.getsize('compressed_bruce.jpg')  # Größe des komprimierten Bilds in Bytes

# Berechne und gebe das Kompressionsverhältnis aus
compression_ratio = original_size / compressed_size
print(f"Originalgröße: {original_size / 1024:.2f} KB")
print(f"Komprimierte Größe: {compressed_size / 1024:.2f} KB")
print(f"Kompressionsverhältnis: {compression_ratio:.2f}")

# Darstellung von Originalbild und komprimiertem Bild
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(bild)
plt.title("Originalbild")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(compressed_img)
plt.title(f"Komprimiertes Bild (K=16) \n Kompressionsverhältnis: {compression_ratio:.2f}")
plt.axis('off')

plt.show()
````

### Auswertung der Kompression
Speichere das komprimierte Bild und berechne
das Kompressionsverhältnis. Vergleiche dazu die Dateigröße des Orginalbildes mit der 
des komprimierten Bildes.

````python
# Berechne und gebe das Kompressionsverhältnis aus
compression_ratio = original_size / compressed_size
print(f"Kompressionsverhältnis: {compression_ratio:.2f}")

# Verlustbehaftete Kompression: Vergleich zwischen Original und komprimiertem Bild
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(bild)
plt.title("Originalbild")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(compressed_img)
plt.title(f"Komprimiertes Bild (K=16) \n Kompressionsverhältnis: {compression_ratio:.2f}")
plt.axis('off')

plt.show()

````

Das Auftauchen eines Hais im Korralenriff hat die Clownfische in Angst und

###  Bestimme ```K```
Während Dorie auf der Suche nach Nemo ist, versuchst 
du die Qualität der Kompression zu verbessern. Dazu bestimmst du den optimalen Wert für ```K``` mit Hilfe der 
Ellenbogenmethode. 

Führe dazu K-Means auf dem Bild ```clownfish.jpg``` mit ```max_iter = 30``` und den
folgenden Werten für ```K``` aus: ```[4, 8, 16, 32, 64, 128, 256]```. Plotte 
die Ellenbogenfunktion und wähle ```K``` aus. Verwende ```score()``` zum Berechnen der Kosten. (Dies kann einige Minuten dauern.)

````python
Ks = [4, 8, 16, 32, 64, 128, 256]
scores = []

for K in Ks:
    kmeans = KMeans(n_clusters=K, max_iter=30)
    kmeans.fit(bildArr_clownfish)
    scores.append(kmeans.score(bildArr_clownfish))

# Ellenbogenplot
plt.plot(Ks, scores, marker='o')
plt.xlabel('Anzahl der Cluster (K)')
plt.ylabel('Kostenfunktion')
plt.title('Ellenbogenmethode')
plt.show()
````
