Die Korrelation lässt sich mit der folgenden Formel berechnen:
    $$r_{X,Y} = {cov(X,Y) \over \sigma_x \sigma_y}$$
    
* $cov_{x,y}$ ist die Kovarianz
* $\sigma_x$ ist die Standardabweichung von X
* $\sigma_y$ ist die Standardabweichung von Y

### Kovarianz:
$$cov_{X,Y} = {\sum_{i}^{N} (x_i - \bar x)(y_i - \bar y) \over N-1}$$
$\bar x$ = Mittelwert

### Standardabweichung:
$$\sigma = \sqrt{{1\over N-1}{\sum_{i}^{N}(x-\bar x)^2}}$$
